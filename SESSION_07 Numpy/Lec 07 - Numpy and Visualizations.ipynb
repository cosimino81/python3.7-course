## First example

#### Line plot

```python
# Creating the array
x = np.arange(2, 21)
y = 2 * x + 5
```

```python
# Set the figure size
plt.figure(figsize=(8,4))

# Plotting the data
plt.plot(x, y, 'b')

# Setting the title
plt.title('Matplot & Numpy Demo')

# Setting the axis names
plt.xlabel('Samples')
plt.ylabel('Values')
plt.show()
```

#### Variation of line plot 

```python
# Set the figure size
plt.figure(figsize=(8,4))

# Plotting the data
plt.plot(x, y, 'ob') #added new parameter

# Setting the title
plt.title('Matplot & Numpy Demo')

# Setting the axis names
plt.xlabel('Samples')
plt.ylabel('Values')
plt.show()
```

## Second Example

#### Exponential plot

```python
# Calculate 2**x for all elements in the array.
x = np.arange(1,11)
y = np.exp2(x)
```

```python
# Make a new plot
plt.figure(figsize=(8,4))
plt.plot(x, y, 'og')
plt.title('Exponential Plot')
plt.xlabel('Samples')
plt.ylabel('Exponential')
plt.show()
```

#### Variationof the exponential plot (bar chart) 

```python
# Make a new plot
plt.figure(figsize=(8,4))
plt.bar(x, y)
plt.title('Exponential Plot')
plt.xlabel('Samples')
plt.ylabel('Exponential')
plt.ylim(-10, 1100)
plt.show()
```

## Third example

#### Histogram

```python
# Build a vector of 10000 normal deviates with variance 0.5^2 and mean 2
mu, sigma = 0, 2
v = np.random.normal(mu,sigma,10000)

# Plot a normalized histogram with 50 bins
plt.hist(v, bins=100, density=1) # matplotlib version (plot)
plt.show()
```

#### Histogram with using seaborn library

```python
# Using seaborn
import seaborn as sns
sns.set_style('whitegrid')

# Build a vector of 10000 normal deviates with variance 0.5^2 and mean 2
mu, sigma = 0, 2
v = np.random.normal(mu,sigma,10000)

sns.distplot(v, bins=100, color='green')
plt.show()
```
